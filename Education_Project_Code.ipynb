```python
import pandas as pd

performance_factors_df = pd.read_csv('/content/StudentPerformanceFactors.csv')

university_data_df = pd.read_csv('/content/IPEDS_data.csv')


```

Visualizations:

```python
import plotly.express as px
import plotly.io as pio
fh = open('visual1.html', 'w')
fh = open('visual2.html', 'w')
fh = open('visual3.html', 'w')
fh = open('visual4.html', 'w')
fh = open('visual5.html', 'w')
fh = open('visual6.html', 'w')
fh = open('visual7.html', 'w')
fh = open('visual8.html', 'w')
fh = open('visual9.html', 'w')
fh = open('visual10.html', 'w')
fh = open('visual11.html', 'w')
fh = open('visual12.html', 'w')

 # this just creates a blank file, so that the write_html() function can write to this file later
fh.close()
index_num=[]
for i in range(len(university_data_df["Name"])):
  index_num.append(1)
last_index=len(university_data_df)
university_data_df["Index"]=index_num

average_cost=[]

size_category = []


enrolled_total=[]
for i in range(len(university_data_df["Total  enrollment"])):
  if i != 257 and i != 1308:
    enrolled_total.append(university_data_df["Total  enrollment"][i])
  else:
    enrolled_total.append(1)

for i in range(len(university_data_df['Total price for out-of-state students living on campus 2013-14'])):

  average_cost.append((university_data_df['Total price for out-of-state students living on campus 2013-14'][i]+university_data_df['Total price for in-state students living on campus 2013-14'][i])/2)


  if university_data_df['Total  enrollment'][i] <=2000:
    size_category.append("small")
  elif university_data_df['Total  enrollment'][i] <= 10000:
    size_category.append("medium")
  else:
    size_category.append("large")




fig1 = px.box(performance_factors_df, x="Family_Income", y="Exam_Score", title = "Student Performance by Socioeconomic Status",
              labels={"Family_Income":"Family Income","Exam_Score":"Exam Score"})
fig1.show()
print("""
Insights:
1. The median score across all different categories of income is roughly the same, with high income students scoring 1 point
more than the other two groups.
2. Although the median score was relatively the same, both the maximum and minimum score for high income students
was the greatest out of all the groups.
3. The low income students had a wider range of scores below the median than the other two groups.
""")
pio.write_html(fig1, file='visual1.html', auto_open=True, full_html=False)


fig2 = px.scatter(performance_factors_df, x='Attendance', y='Exam_Score', title='Attendance Rate vs Performance',
labels={'Attendance': 'Attendance Rate (%)','Exam_Score': 'Exam Score', 'Hours_Studied':'Hours Studied','Previous_Scores':"Previous Score"}, trendline="ols", color = 'Hours_Studied',
                  hover_data='Previous_Scores')
fig2.show()
print("""
Insights:
1. As attendance rate increases, so do test scores, on average. There are a few outliers but for the most part,
those who attended more classes scored higher.
2. There is also a trend in which those who studied longer performed better. Even if their attendance was poor, longer studying
times still tended to improve scores, especially when compared with other students with similar attendance rates.
3. Students who had good attendance and studied more tended to score closer to or better than their previous scores on this exam,
while students that studied less or had lower attendance tended to score much worse than they had in the past.
""")

pio.write_html(fig2, file='visual2.html', auto_open=True, full_html=False)


fig3 = px.scatter(university_data_df, x=average_cost,
                  y=["Bachelor's degrees awarded", "Master's degrees awarded"], title='Tuition and Fees (2013-14) vs Degrees Awarded',
                  labels={'x': 'Average Cost (2013-2014) ($)',"value": 'Degrees Awarded',"variable":"Degree Type"}, hover_name='Name', hover_data="Total  enrollment",
                  facet_col=size_category, trendline='ols', log_y=True, trendline_color_override="black")
fig3.show()
pio.write_html(fig3, file='visual3.html', auto_open=True, full_html=False)
print("""
Insights:
1. As cost increases, the total amount of degrees awarded tends to remain relatively steady on average.
2. As the cost increases for small schools, both bachelor's and master's degrees increase, but as the cost increases
for medium and large schools, bachelor's degrees tend to slightly decrease and master's degrees tend to increase.
3. At large, expensive schools, the number of bachelor's degrees and master's degrees awarded are very similar,
but at expensive medium and small schools, there are still many more bachelor's degrees awarded than masters degrees.
""")



fig4 = px.scatter_geo(university_data_df, lat='Latitude location of institution', lon='Longitude location of institution', color='ACT Composite 75th percentile score',
hover_name='Name', title='University Size & ACT Scores Across the US',size=enrolled_total)
fig4.update_layout(geo_scope='usa')
fig4.show()
pio.write_html(fig4, file='visual4.html', auto_open=True, full_html=False)
print("""
Insights:
1. There are many more universities in the eastern half of the US than the western, with an especially large concentration
in the Northeastern coastal region.
2. Larger universities tend to have higher ACT scores than the small universities do. There are few large universities
with a score of 20 or under
3. There are no real distinctions that can be drawn regarding education quality based on the graph. No geographical region
has a clearly higher or lower score average than others, and the most noticeable distinction is the location of the schools.
""")

fig5 = px.parallel_categories(performance_factors_df,dimensions=['Motivation_Level', 'Parental_Involvement', 'Peer_Influence', 'Internet_Access'],
title='Motivation vs Parental Involvement vs Peer Influence vs Internet Access')
fig5.show()
print("""
Insights:
1. The vast majority of students have access to the internet.
""")
pio.write_html(fig5, file='visual5.html', auto_open=True, full_html=False)



fig6 = px.box(performance_factors_df, x="Sleep_Hours",y="Exam_Score", labels={"Sleep_Hours":"Hours of Sleep","Exam_Score":"Exam Score","Hours_Studied":"Hours Studied"},
              title="Hours of Sleep versus Exam Score")
fig6.show()
pio.write_html(fig6, file='visual6.html', auto_open=True, full_html=False)
print("""
Insights:
1. Students who got 7 hours of sleep tended to study the most out of all student groups.




""")




fig7 = px.scatter(performance_factors_df, x="Hours_Studied", y="Exam_Score", title = "Student Performance by Parental Education", color="Parental_Education_Level",
labels={'Exam_Score':'Exam Score',"Hours_Studied":"Hours Studied"},
text = "Exam_Score")
fig7.show()
pio.write_html(fig7, file='visual7.html', auto_open=True, full_html=False)
print("""
Insights:
1. The majority of students
""")





fig8 = px.box(performance_factors_df, x="Motivation_Level", y="Exam_Score", facet_col="Peer_Influence",
                labels={'Motivation_Level':'Motivation Level',"Exam_Score":'Exam Score'}, title="Motivation Level versus Hours Studied compared with Peer Influence")
pio.write_html(fig8, file='visual8.html', auto_open=True, full_html=False)
fig8.show()


fig9 = px.scatter_geo(university_data_df, lat='Latitude location of institution', lon='Longitude location of institution', color='Religious affiliation',
hover_name='Name',hover_data="Sector of institution", title='Religious affiliation of Universities across the US',size=enrolled_total)
fig9.update_layout(geo_scope='usa')
fig9.show()
pio.write_html(fig9, file='visual9.html', auto_open=True, full_html=False)

fig10=px.box(performance_factors_df, x="Parental_Involvement",y="Exam_Score", labels={"Parental_Involvement":"Parental Involvement Level","Exam_Score":"Exam Score","Tutoring_Sessions":"Tutoring Sessions"}, title="Parental Involvement versus Exam Score",
             color="Tutoring_Sessions")
fig10.show()
pio.write_html(fig10, file='visual10.html', auto_open=True, full_html=False)

fig11=px.pie(university_data_df, values = "Index", names="Historically Black College or University", title="Historically Black Colleges or Universities in the United States")
pio.write_html(fig11, file='visual11.html', auto_open=True, full_html=False)
fig11.show()


fig12=px.pie(university_data_df, values="Index",names="Degree of urbanization (Urban-centric locale)", title= "Degree of Urbanization of Colleges and Universities in the US")
fig12.show()
pio.write_html(fig12, file='visual12.html', auto_open=True, full_html=False)
```
